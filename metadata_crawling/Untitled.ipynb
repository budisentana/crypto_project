```python
import pandas as pd
df_comment = pd.read_csv('/home/budi/crypto_project/crypto_code/apps_screening/comment_file.csv',low_memory=False)
df_comment['star'] = df_comment['star'].str.strip()
df_comment
```

```python

# remove error
# len(df_comment[df_comment['star']=='error'])
df_comment= df_comment[~df_comment['star'].isin(['error'])]

df_comment
```

```python
#find apps containing comment
df_all = df_comment.groupby(['appId']).size().reset_index(name='count')
df_all
```

```python
# convert rating to int

df_comment['star'] = df_comment['star'].astype(float)
df_comment
```

```python
# df_comment['star'] = df_comment['star'].astype(int)
positive = df_comment[df_comment['star']>3]
positive
```

```python
positive_app = positive.groupby(['appId']).size().reset_index()
positive_app
```
