```python
perm_sum_path ='/home/budi/crypto_project/crypto_code/static_analysis/report/permission_sum_per_app.csv'

import pandas as pd

df_perm = pd.read_csv(perm_sum_path)
df_perm = df_perm.fillna(0)
df_perm['total'] = df_perm['customized']+ df_perm['dangerous']+df_perm['normal']+df_perm['signature']
df_perm= df_perm.sort_values(by=['total'],ascending=False)
df_perm
```

```python
df_lest = df_perm[df_perm['total']<2]
df_lest
```

```python
df_avg = df_perm[['total','customized','dangerous','normal','signature']].agg(['mean']).unstack()
df_avg
```

```python
df_cust= df_perm[df_perm['customized']>0]
df_cust
```

```python
detail_path = '/home/budi/crypto_project/crypto_code/static_analysis/report/permission_detail_fixing.csv'
df_detail = pd.read_csv(detail_path)
df_detail
```

```python
df_cust_detail = df_detail[df_detail['level']=='customized']
df_cust_detail
```

```python
df_cust_group = df_cust_detail.groupby(['permission'])['permission'].count().reset_index(name='count').sort_values(by=['count'],ascending=False)
df_cust_group
```

```python
df_cust_group100 = df_cust_group[df_cust_group['count']>=100]
df_cust_group100
```

```python
df_badge = df_detail[df_detail['permission'].str.contains("BADGE", case=False)]
df_badge.to_csv('/home/budi/crypto_project/crypto_code/static_analysis/report/permission_BADGE.csv')
df_badge.head(100)
```

```python
df_badge_count = df_badge.groupby(['app_id'])['app_id'].count().reset_index(name='count').sort_values(by=['count'],ascending=False)
df_badge_count_large = df_badge_count[df_badge_count['count']>5]
df_badge_count_large
```
